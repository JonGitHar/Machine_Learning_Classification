```python
# Import modules for data ingestion and manipulation

import pandas as pd
import numpy as np

# Import csv from previous notebook as a data frame

df_trim_3 = pd.read_csv("applications_2.csv", index_col = 0)
df_trim_3.info()
```

```python
df_trim_3
```

```python
df_trim_3.isnull().sum()
```

```python
# Let's drop some more columns that we intuitively think may not be interpretable,
# may have too many categories, and/or may be correlated with other existing features

df_trim_3.drop(["EMPLOYER_YEAR_COMMENCED_BUSINESS", "JOB_TITLE", "MINIMUM_EDUCATION", 
                "MAJOR_FIELD_OF_STUDY.1", "EMPLOYER_YEAR_COMMENCED_BUSINESS", "REQUIRED_EXPERIENCE", 
                "PROFESSIONAL_OCCUPATION"], axis = 1, inplace = True)
```

```python
df_trim_3["MAJOR_FIELD_OF_STUDY"].value_counts().head(15)
```

```python
# Convert "MAJOR_FIELD_OF_STUDY" into additional binary columns

COMP_SCI_AND_ENGI = ["Computer Science", "COMPUTER SCIENCE", "Comp Sci, MIS, Engg any, Math, Bus Admin,or rel quantitative field", 
                     "Electrical Engineering", "Mechanical Engineering", "Computer Engineering", "Engineering",
                     "Computer Science, Engineering, Business, Technology or related STEMquantitative field", 
                     "Comp Sci, MIS, Engg (any), Math, Bus Admin,or rel quantitative field",
                     "Eng Any, ComSci, ComApp, BusAdm, BusFin, BusAcc, BusMgmt, ComInfoSys, MgmtInfoSys, Math or rel"]
df_trim_3["MAJOR_IS_COMP_SCI_OR_ENGINEERING"] = np.where(df_trim_3["MAJOR_FIELD_OF_STUDY"].isin(COMP_SCI_AND_ENGI), 1, 0)
df_trim_3["MAJOR_IS_MEDICINE"] = np.where(df_trim_3["MAJOR_FIELD_OF_STUDY"] == "Medicine", 1, 0)
df_trim_3["MAJOR_IS_ACCOUNTING"] = np.where(df_trim_3["MAJOR_FIELD_OF_STUDY"] == "Accounting", 1, 0)
df_trim_3["MAJOR_IS_BUSINESS_ADMIN"] = np.where(df_trim_3["MAJOR_FIELD_OF_STUDY"] == "Business Administration", 1, 0)

# Drop "MAJOR_FIELD_OF_STUDY" column as it is no longer needed

df_trim_3.drop(["MAJOR_FIELD_OF_STUDY"], axis = 1, inplace = True)
```

```python
# Convert "FOREIGN_WORKER_ED_INST_COUNTRY" into a new binary column to indicate whether or not 
# Applicant received education in the US

df_trim_3["RECEIVED_EDUCATION_IN_USA"] = np.where(df_trim_3["FOREIGN_WORKER_ED_INST_COUNTRY"] == "UNITED STATES OF AMERICA", 1, 0)
df_trim_3.drop(["FOREIGN_WORKER_ED_INST_COUNTRY"], axis = 1, inplace = True)
```

```python
df_trim_3.groupby("FOREIGN_WORKER_EDUCATION").count()
```

```python
# Convert "FOREIGN_WORKER_EDUCATION" into additional binary columns
# We assume that a Doctorate applicant has both a Master's and Bachelor's
# We also assume that a Master's applicant has a Bachelor's

BACHELOR = ["Bachelor's", "Master's", "Doctorate"]
MASTER = ["Master's", "Doctorate"]
df_trim_3["POSSESSES_BACHELORS_DEGREE"] = np.where(df_trim_3["FOREIGN_WORKER_EDUCATION"].isin(BACHELOR), 1, 0)
df_trim_3["POSSESSES_MASTERS_DEGREE"] = np.where(df_trim_3["FOREIGN_WORKER_EDUCATION"].isin(MASTER), 1, 0)
df_trim_3["POSSESSES_DOCTORATE_DEGREE"] = np.where(df_trim_3["FOREIGN_WORKER_EDUCATION"] == "Doctorate", 1, 0)


# Drop "FOREIGN_WORKER_EDUCATION" column as it is no longer needed

df_trim_3.drop(["FOREIGN_WORKER_EDUCATION"], axis = 1, inplace = True)
```

```python
# Convert "CLASS_OF_ADMISSION" into binary columns to indicate if applicant was admitted under 
# H1B (specialized skills) visa or TN (Mexican/Canadian) visa 

df_trim_3["H1B_VISA_HOLDER"] = np.where(df_trim_3["CLASS_OF_ADMISSION"] == "H-1B", 1, 0)
df_trim_3["TN_VISA_HOLDER"] = np.where(df_trim_3["CLASS_OF_ADMISSION"] == "TN", 1, 0)

# Drop "CLASS_OF_ADMISSION" column as it is no longer needed

df_trim_3.drop(["CLASS_OF_ADMISSION"], axis = 1, inplace = True)
```

```python
df_trim_3.isnull().sum()
```

```python
# Final column we will work with is "COUNTRY_OF_CITIZENSHIP"
# We will merge another data set onto the existing data frame using this column value

df_trim_3["COUNTRY_OF_CITIZENSHIP"].value_counts().head(30)
```

```python
# We will import educational indicators from the World Bank Data website 
# https://data.worldbank.org/indicator
# Of particular interest are the country's 
# 1) literacy rate 
# 2) govt expenditure on education as % of GDP
# 3) R&D researchers per million people

NATL_LITERACY_RATE = pd.read_csv("Literacy_rate.csv")
NATL_EDU_EXPENDITURE_PCT_GDP = pd.read_csv("Education_Expenditure.csv")
NATL_RD_PER_MILLION = pd.read_csv("RD researchers per million.csv")
```

```python
NATL_LITERACY_RATE
```

```python
NATL_EDU_EXPENDITURE_PCT_GDP
```

```python
# Check how many countries have entries of 0 (not available) in each of the three new dataframes

print ((NATL_LITERACY_RATE.values == 0).sum())
print ((NATL_EDU_EXPENDITURE_PCT_GDP.values == 0).sum())
print ((NATL_RD_PER_MILLION.values == 0).sum())
```

```python
# We will disregard the data for RD researchers per million as there are too many countries without available metric
# For the others, perform a left merge with the main data frame on the left
# First, prepare the data for merging

NATL_LITERACY_RATE["Country"] = NATL_LITERACY_RATE["Country"].str.upper()
NATL_EDU_EXPENDITURE_PCT_GDP["Country"] = NATL_EDU_EXPENDITURE_PCT_GDP["Country"].str.upper()

# Also check which countries from main data frame do not match the ones from the new data frames

MAIN_COUNTRY_LIST = df_trim_3["COUNTRY_OF_CITIZENSHIP"].unique()
NEW_COUNTRY_LIST = NATL_LITERACY_RATE["Country"].unique()
list_difference = [item for item in MAIN_COUNTRY_LIST if item not in NEW_COUNTRY_LIST]
print(list_difference)
```

```python
# Manually change country names the two new data frames to resolve the difference

replace_values = {"KOREA, REP." : "SOUTH KOREA", "IRAN, ISLAMIC REP.": "IRAN", "VENEZUELA, RB": "VENEZUELA", 
                  "UNITED STATES": "UNITED STATES OF AMERICA", "MACAO SAR, CHINA": "MACAU",  
                  "RUSSIAN FEDERATION": "RUSSIA", "NORTH MACEDONIA": "MACEDONIA", "HONG KONG SAR, CHINA": "HONG KONG", 
                  "EGYPT, ARAB REP.": "EGYPT", "BAHAMAS, THE": "BAHAMAS", "SYRIAN ARAB REPUBLIC": "SYRIA", 
                  "CONGO, DEM. REP.": "DEMOCRATIC REPUBLIC OF CONGO", "COTE D'IVOIRE": "COTE d'IVOIRE", 
                  "MYANMAR": "BURMA (MYANMAR)", "ST. LUCIA": "ST LUCIA", "ST. KITTS AND NEVIS": "ST KITTS AND NEVIS",
                  "LAO PDR": "LAOS", "GAMBIA, THE": "GAMBIA", "SLOVAK REPUBLIC": "SLOVAKIA", 
                  "KYRGYZ REPUBLIC": "KYRGYZSTAN", "MICRONESIA, FED. STS.": "MICRONESIA", "WEST BANK AND GAZA": "PALESTINE",
                  "ESWATINI": "SWAZILAND", "CONGO, REP.": "REPUBLIC OF CONGO", "SERBIA": "SERBIA AND MONTENEGRO",
                  "BRUNEI DARUSSALAM": "BRUNEI", "YEMEN, REP.": "YEMEN", "ST. VINCENT AND THE GRENADINES":
                  "SAINT VINCENT AND THE GRENADINES", "CABO VERDE": "CAPE VERDE"} 

df_trim_3["COUNTRY_OF_CITIZENSHIP"].replace("IVORY COAST", "COTE d'IVOIRE", inplace = True)
df_trim_3["COUNTRY_OF_CITIZENSHIP"].replace("SOVIET UNION", "RUSSIA", inplace = True)
df_trim_3["COUNTRY_OF_CITIZENSHIP"].replace("PALESTINIAN TERRITORIES", "PALESTINE", inplace = True)
df_trim_3["COUNTRY_OF_CITIZENSHIP"].replace("TAIWAN", "CHINA", inplace = True)
df_trim_3["COUNTRY_OF_CITIZENSHIP"].replace("SERBIA", "SERBIA AND MONTENEGRO", inplace = True)

NATL_LITERACY_RATE = NATL_LITERACY_RATE.replace({"Country": replace_values})
NATL_EDU_EXPENDITURE_PCT_GDP = NATL_EDU_EXPENDITURE_PCT_GDP.replace({"Country": replace_values})
```

```python
MAIN_COUNTRY_LIST = df_trim_3["COUNTRY_OF_CITIZENSHIP"].unique()
NEW_COUNTRY_LIST = NATL_RD_PER_MILLION["Country"].unique()
list_difference = [item for item in MAIN_COUNTRY_LIST if item not in NEW_COUNTRY_LIST]
print(list_difference)
```

```python
# Proceed with merging

df4 = pd.merge(df_trim_3, NATL_LITERACY_RATE, left_on = ["COUNTRY_OF_CITIZENSHIP"], right_on = ["Country"], how = "left")
df5 = pd.merge(df4, NATL_EDU_EXPENDITURE_PCT_GDP, left_on = ["COUNTRY_OF_CITIZENSHIP"], right_on = ["Country"], how = "left")
```

```python
# Check how many null entries we have following the merge

df5.isnull().sum()
```

```python
# Drop the unnecessary columns for the final time

df5.drop(["COUNTRY_OF_CITIZENSHIP", "Country_x", "Country_y"], axis = 1, inplace = True)
```

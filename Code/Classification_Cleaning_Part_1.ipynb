```python
# Import modules for data ingestion and manipulation

import pandas as pd
import numpy as np
```

```python
# Pull in the Q2 2021 data as well as the entirety of 2020 data
# Unfortunately, Q1 2021 cannot be obtained at current time from any source

df_2020 = pd.read_excel("PERM_Disclosure_Data_FY2020.xlsx")
df_2021_q2 = pd.read_excel("PERM_Disclosure_Data_FY2021_Q2.xlsx")
```

```python
# As the data frame retained the same format and column namings heading from 2020 into 2021, 
# we can concatenate the two data sets

df = pd.concat([df_2020, df_2021_q2])
```

```python
df.info()
```

```python
# Inspect the breakdown of case status/results 

df.groupby("CASE_STATUS").count()
```

```python
# We can trim the data frame by only considering entries that were either certified(approved) or denied

df_trim_1 = df.loc[df["CASE_STATUS"].isin(["Certified", "Denied"])]
df_trim_1.info()
```

```python
# "CASE_STATUS" column is our target variable. Certified will be redesignated as "1" and Denied will be redesignated as "0"

df_trim_1.loc[df_trim_1["CASE_STATUS"] == "Certified", "CASE_STATUS"] = 1
df_trim_1.loc[df_trim_1["CASE_STATUS"] == "Denied", "CASE_STATUS"] = 0
df_trim_1.groupby("CASE_STATUS").count()
```

```python
# List all columns so we select a small subset of the features/columns to perform baseline analysis and modeling

list(df_trim_1.columns.values)
```

```python
# Only select columns that may prove useful

df_trim_2 = df_trim_1[['CASE_STATUS', 'RECEIVED_DATE', 'DECISION_DATE', 'EMPLOYER_NAME', 
                       'EMPLOYER_COUNTRY', 'EMPLOYER_NUM_EMPLOYEES','EMPLOYER_YEAR_COMMENCED_BUSINESS', 
                       'PW_SOC_TITLE', 'PW_SKILL_LEVEL', 'PW_WAGE', 'PW_UNIT_OF_PAY', 'WAGE_OFFER_FROM', 
                       'WAGE_OFFER_TO', 'WAGE_OFFER_UNIT_OF_PAY', 'JOB_TITLE', 'MINIMUM_EDUCATION', 'MAJOR_FIELD_OF_STUDY',
                       'SPECIFIC_SKILLS', 'MAJOR_FIELD_OF_STUDY', 'REQUIRED_EXPERIENCE', 'PROFESSIONAL_OCCUPATION', 
                       'COUNTRY_OF_CITIZENSHIP', 'FOREIGN_WORKER_BIRTH_COUNTRY', 'CLASS_OF_ADMISSION', 'FOREIGN_WORKER_EDUCATION',
                       'FOREIGN_WORKER_INFO_MAJOR', 'FOREIGN_WORKER_INST_OF_ED', 'FOREIGN_WORKER_ED_INST_COUNTRY', 'EMP_INFO_DECL_TITLE']]
```

```python
# Save current data frame to CSV and continue cleaning in next notebook

df_trim_2.to_csv("applications.csv")
```

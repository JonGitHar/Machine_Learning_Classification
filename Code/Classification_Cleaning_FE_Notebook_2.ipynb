```python
# Import modules for data ingestion and manipulation

import pandas as pd
import numpy as np

# Import csv from previous notebook as a data frame

df_trim_2 = pd.read_csv("applications.csv", index_col = 0)
df_trim_2.info()
```

```python
df_trim_2.head()
```

```python
# Check breakdown by countries and whether or not the application was approved
# We can see that the top 5 sources of approved applications are from India, China, Mexico, Canada, and South Korea

df_trim_2.groupby(["COUNTRY_OF_CITIZENSHIP", "CASE_STATUS"]).count().sort_values(["RECEIVED_DATE"], ascending = False)
```

```python
# Let's further tweak the columns "RECEIVED_DATE" and "DECISION_DATE"
# We want to learn - 
# 1) if the decision date was during the Trump or Biden (sworn in on January 20, 2021) presidency
# 2) Days elapsed between date of application receipt and decision

# Prep these two columns first by taking trimming the input values and converting them to datetime format

df_trim_2["RECEIVED_DATE"] = df_trim_2["RECEIVED_DATE"].str[:10]
df_trim_2["DECISION_DATE"] = df_trim_2["DECISION_DATE"].str[:10]
df_trim_2["RECEIVED_DATE"] = pd.to_datetime(df_trim_2["RECEIVED_DATE"])
df_trim_2["DECISION_DATE"] = pd.to_datetime(df_trim_2["DECISION_DATE"])
```

```python
# Create additional columns to -
# 1) See days elapsed between received date and decision date
# 2) Check if decision date was during Trump or Biden presidency

df_trim_2["TRUMP_PRESIDENCY"] = np.where(df_trim_2["DECISION_DATE"] < "2021-01-20", 1, 0)
df_trim_2["BIDEN_PRESIDENCY"] = np.where(df_trim_2["TRUMP_PRESIDENCY"] == 1, 0, 1)
df_trim_2["PROCESSING_TIME_IN_DAYS"] = df_trim_2["DECISION_DATE"] - df_trim_2["RECEIVED_DATE"]
df_trim_2["PROCESSING_TIME_IN_DAYS"] = df_trim_2["PROCESSING_TIME_IN_DAYS"].astype(str)
df_trim_2["PROCESSING_TIME_IN_DAYS"] = df_trim_2["PROCESSING_TIME_IN_DAYS"].str[:-4]
df_trim_2["PROCESSING_TIME_IN_DAYS"] = df_trim_2["PROCESSING_TIME_IN_DAYS"].astype(int)

# Let's also remove the received date and decision date columns as we no longer need them
df_trim_2.drop(df_trim_2.columns[[1, 2]], axis = 1, inplace = True)
df_trim_2
```

```python
# Let's also see the employer breakdown

df_trim_2.groupby("EMPLOYER_COUNTRY").count()
```

```python
# As all but two applicants have employers based out of the USA, we can remove this employer country column as well

df_trim_2.drop(["EMPLOYER_COUNTRY"], axis = 1, inplace = True)
```

```python
# Let's also drop some additional columns that we intuitively think may not be interpretable,
# may have too many categories, and/or may be correlated with other existing features

df_trim_2.drop(["FOREIGN_WORKER_INFO_MAJOR", "FOREIGN_WORKER_BIRTH_COUNTRY", "FOREIGN_WORKER_INST_OF_ED", 
                "EMP_INFO_DECL_TITLE", "PW_SOC_TITLE", "PW_WAGE", "PW_UNIT_OF_PAY"], axis = 1, inplace = True)
df_trim_2.info()
```

```python
# Quickly check number of empty entries in each column

df_trim_2.isnull().sum()
```

```python
# Drop any rows without values in the "WAGE_OFFER_FROM" and other columns of interest

df_trim_2.dropna(subset = ["WAGE_OFFER_FROM", "WAGE_OFFER_UNIT_OF_PAY", "FOREIGN_WORKER_EDUCATION", "MINIMUM_EDUCATION", "REQUIRED_EXPERIENCE", "COUNTRY_OF_CITIZENSHIP"], inplace = True)

# Impute missing values from "EMPLOYER_NUM_EMPLOYEES" column with the mean value

df_trim_2["EMPLOYER_NUM_EMPLOYEES"].fillna(df_trim_2["EMPLOYER_NUM_EMPLOYEES"].mean(), inplace = True)

# If anything is listed under "SPECIFIC_SKILLS", change value to 1. If left blank, fill with 0.
df_trim_2["SPECIFIC_SKILLS"] = df_trim_2["SPECIFIC_SKILLS"].fillna(0)
df_trim_2["SPECIFIC_SKILLS"] = np.where(df_trim_2["SPECIFIC_SKILLS"] != 0, 1, 0)

# Combine wage "WAGE_OFFER_FROM" and "WAGE_OFFER_TO" into a single column by taking their average if both columns are populated
# For rows where only "WAGE_OFFER_FROM" is populated, set "WAGE_OFFER_TO" equal to "WAGE_OFFER_FROM"

df_trim_2["WAGE_OFFER_TO"].fillna(df_trim_2["WAGE_OFFER_FROM"], inplace = True)
df_trim_2["WAGE_PER_PAY_PERIOD"] = (df_trim_2["WAGE_OFFER_FROM"] + df_trim_2["WAGE_OFFER_TO"]) / 2
df_trim_2.drop(["WAGE_OFFER_FROM", "WAGE_OFFER_TO"], axis = 1, inplace = True)
```

```python
# We see below that not all wages are in annual salary terms

df_trim_2.groupby("WAGE_OFFER_UNIT_OF_PAY").count()
```

```python
df_trim_2.isnull().sum()
```

```python
# Let's transform the wage offer unit of pay into number of pay periods per year, then multiply by wage per pay period

df_trim_2["WAGE_OFFER_UNIT_OF_PAY"] = np.select([df_trim_2["WAGE_OFFER_UNIT_OF_PAY"] == "Bi-Weekly", 
                                                 df_trim_2["WAGE_OFFER_UNIT_OF_PAY"] == "Hour", 
                                                 df_trim_2["WAGE_OFFER_UNIT_OF_PAY"] == "Month", 
                                                 df_trim_2["WAGE_OFFER_UNIT_OF_PAY"] == "Week", 
                                                 df_trim_2["WAGE_OFFER_UNIT_OF_PAY"] == "Year"],[26, 2080, 12, 52, 1])
df_trim_2["ANNUAL_SALARY_OFFER"] = df_trim_2["WAGE_PER_PAY_PERIOD"] * df_trim_2["WAGE_OFFER_UNIT_OF_PAY"]
df_trim_2["ANNUAL_SALARY_OFFER"] = df_trim_2["ANNUAL_SALARY_OFFER"].astype(int)
df_trim_2.drop(["WAGE_PER_PAY_PERIOD", "WAGE_OFFER_UNIT_OF_PAY"], axis = 1, inplace = True)
```

```python
df_trim_2.groupby("PW_SKILL_LEVEL").count()
```

```python
# Change the "PW_SKILL_LEVEL" column values to a range of numbers from 1 to 4 and 
# Consider missing values as 0

df_trim_2["PW_SKILL_LEVEL"] = np.select([df_trim_2["PW_SKILL_LEVEL"] == "Level I", df_trim_2["PW_SKILL_LEVEL"] == "Level II", 
                                         df_trim_2["PW_SKILL_LEVEL"] == "Level III", df_trim_2["PW_SKILL_LEVEL"] == "Level IV"], 
                                         [1, 2, 3, 4])
```

```python
# Check which employers have the most filings

df_trim_2["EMPLOYER_NAME"].value_counts().head(15)
```

```python
# Let's use the top 10 entries above along with FAANG (Facebook, Amazon, Apple, Netflix, Google)
# To create a new feature column indicating whether or not applicant belongs to those employers

TOP_EMPLOYERS = ["MICROSOFT CORPORATION", "GOOGLE LLC", "TATA CONSULTANCY SERVICES LIMITED", "INFOSYS LTD.",
                 "FACEBOOK, INC.", "INTEL CORPORATION", "AMAZON.COM SERVICES, INC.", "AMAZON.COM SERVICES LLC",
                 "APPLE INC.", "CAPGEMINI AMERICA, INC.", "NETFLIX", "Netflix, Inc."]
df_trim_2["FAANG_AND_OTHER_PROMINENT_EMPLOYERS"] = np.where(df_trim_2["EMPLOYER_NAME"].isin(TOP_EMPLOYERS), 1, 0)
df_trim_2.drop(["EMPLOYER_NAME"], axis = 1, inplace = True)
```
